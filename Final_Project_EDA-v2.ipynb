```
##### Import libraries ########
import pandas as pd # Data manipulation
import numpy as np # Data manipulation
import seaborn as sns # Visualization
import matplotlib.pyplot as plt # Visualization
import sweetviz as sv # Basic EDA 
sns.set(style="darkgrid") # seaborn style
```

```
kick_base = pd.read_csv('ks-projects-201801.csv', sep=',', error_bad_lines=False)
kick_base.head(3)
```

```
# Create feature for pledged/goal percentage
kick_base['percent_pledged'] = kick_base['usd_pledged_real']/kick_base['usd_goal_real']
```

```
kick_base_report = sv.analyze(kick_base)
```

```
# Could limit to US both in terms of currency and country
# Launched should be bucketed into year, maybe month 
```

```
# In Colab you have to open this from the files folder
kick_base_report.show_html('Kick_base.html')
```

```
# Mosaic plot 
plt.rcParams['figure.figsize']=(18,12)
plt.rcParams['font.size'] = 8.0
mosaic(kick_base.sort_values('state'), ['main_category', 'state'], axes_label=True, gap=0.002)
plt.show
```
